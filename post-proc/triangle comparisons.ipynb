```python
import numpy as np
import getdist
import matplotlib.pyplot as plt
from getdist import MCSamples, plots
```

```python
import matplotlib as mpl

mpl.rcParams['axes.linewidth'] = 1.5
```

```python
a1l128_white = np.load('../results/samples_a1l128_white.npy')
a1l128_biased = np.load('../results/samples_a1l128_biased.npy')
a1l128 = np.load('../results/samples_a1l128.npy')
```

```python
param_names = [("param1", r"\alpha"), ("param2", r"\ell_0"), ("param3", r"\beta_d"), ("param4", "T_d"), ("param5", r"\beta_s")]

s1 = MCSamples(samples=a1l128_white, names=[p[0] for p in param_names], labels=[p[1] for p in param_names])
s2 = MCSamples(samples=a1l128_biased, names=[p[0] for p in param_names], labels=[p[1] for p in param_names])
s3 = MCSamples(samples=a1l128, names=[p[0] for p in param_names], labels=[p[1] for p in param_names])
```

```python
g = plots.get_subplot_plotter(width_inch = 8)

g.settings.lab_fontsize = 26         # axis label size
g.settings.linewidth = 2.5         # line width for contours
g.settings.axes_fontsize = 16.5  # or your desired size

g.triangle_plot([s2, s1, s3], ['param1', 'param2'], filled=True, colors = ['red', 'green', 'blue'], line_args=[{"color": "red"},{"color": "green"},{"color": "blue"}],
                legend_labels = ['No Bias Correction', 'Bias Correction - White Noise', r'Bias Correction - Pink Noise'])

x_vals = [-1, 128, 1.54, 20, -3]  # example x positions
y_vals = x_vals  # example y positions

# Access the triangle axes
for ax in g.subplots.flatten():
    if ax is not None:
        for x in x_vals:
            ax.axvline(x, color='k', linestyle='--', linewidth=1.5)
        for y in y_vals:
            ax.axhline(y, color='k', linestyle='--', linewidth=1.5)

plt.savefig('plots_white/a1l128_3_comparison.png')
```
