```python
import numpy as np
import getdist
import matplotlib.pyplot as plt
from getdist import MCSamples, plots
```

```python
import matplotlib as mpl

mpl.rcParams['axes.linewidth'] = 1.5
```

## Variation with $\ell_0$

```python
a1l64 = np.load('../results/samples_a1l64_white.npy')[:, 2:]
a1l128 = np.load('../results/samples_a1l128_white.npy')[:, 2:]
a1l256 = np.load('../results/samples_a1l256_white.npy')[:, 2:]
a1lv = np.load('../results/samples_a1lv_white.npy')[:, -3:]
```

```python
param_names = [ ("param1", r"\beta_d"), ("param2", "T_d"), ("param3", r"\beta_s")]

s0 = MCSamples(samples=a1l64, names=[p[0] for p in param_names], labels=[p[1] for p in param_names], label = r'$\ell_0$ = $\ell_\max$/4')
s1 = MCSamples(samples=a1l128, names=[p[0] for p in param_names], labels=[p[1] for p in param_names], label = r'$\ell_0$ = $\ell_\max$/2')
s2 = MCSamples(samples=a1l256, names=[p[0] for p in param_names], labels=[p[1] for p in param_names], label = r'$\ell_0$ = $\ell_\max$')
s3 = MCSamples(samples=a1lv, names=[p[0] for p in param_names], labels=[p[1] for p in param_names], label = r'$\ell_0$ = Variable')
```

```python
g = plots.get_subplot_plotter(subplot_size = 10,width_inch = 5)

# Set line style default
g.settings.line_styles = 'tab10'
g.settings.axes_fontsize = 15
g.settings.axes_labelsize = 20

g.plot_1d([s0, s1, s2, s3], 'param1', nx=4, legend_ncol=4, lws=[3,3,3,3])
g.add_legend(legend_labels=[r'$\ell_0 = \ell_\max/4$', r'$\ell_0 = \ell_\max/2$', r'$\ell_0 = \ell_\max$', r'$\ell_0$ = Variable'])
plt.axvline(x = 1.54, color = 'k', linestyle = '--', linewidth = 2.5)
# plt.ylabel(r'$\mathcal{L}$', fontsize = 25)
plt.ylabel('Posterior', fontsize = 20)
# plt.title(r'Variation in $\beta_d$', fontsize = 15)
plt.savefig('plots_white/ell0_variation_betad.png', bbox_inches = 'tight')
plt.show()

g = plots.get_subplot_plotter(subplot_size=10, width_inch = 5)

# Set line style default
g.settings.line_styles = 'tab10'
g.settings.axes_fontsize = 15
g.settings.axes_labelsize = 20
g.plot_1d([s0, s1, s2, s3], 'param2', nx=4, legend_ncol=3, lws=[3,3,3,3])
plt.ylabel('Posterior', fontsize = 20)
g.add_legend(legend_labels=[r'$\ell_0 = \ell_\max/4$', r'$\ell_0 = \ell_\max/2$', r'$\ell_0 = \ell_\max$', r'$\ell_0$ = Variable'])
plt.axvline(x = 20, color = 'k', linestyle = '--', linewidth = 2.5)
# plt.title(r'Variation in $T_d$', fontsize = 15)
plt.savefig('plots_white/ell0_variation_Td.png', bbox_inches = 'tight')
plt.show()

g = plots.get_subplot_plotter(subplot_size=10, width_inch = 5)

# Set line style default
g.settings.line_styles = 'tab10'
g.settings.axes_fontsize = 15
g.settings.axes_labelsize = 20
g.plot_1d([s0, s1, s2, s3], 'param3', nx=4, legend_ncol=3, lws=[3,3,3,3])
plt.ylabel('Posterior', fontsize = 20)
g.add_legend(legend_labels=[r'$\ell_0 = \ell_\max/4$', r'$\ell_0 = \ell_\max/2$', r'$\ell_0 = \ell_\max$', r'$\ell_0$ = Variable'])
plt.axvline(x = -3, color = 'k', linestyle = '--', linewidth = 2.5)
# plt.title(r'Variation in $\beta_s$', fontsize = 15)
plt.savefig('plots_white/ell0_variation_betas.png', bbox_inches = 'tight')
plt.show()

```

## Variation with $\alpha$

```python
a0l128 = np.load('../results/samples_a0l128_white.npy')[:, 2:]
a1l128 = np.load('../results/samples_a1l128_white.npy')[:, 2:]
a2l128 = np.load('../results/samples_a2l128_white.npy')[:, 2:]
avl128 = np.load('../results/samples_avl128_white.npy')[:, -3:]
```

```python
param_names = [ ("param1", r"\beta_d"), ("param2", "T_d"), ("param3", r"\beta_s")]

s0 = MCSamples(samples=a0l128, names=[p[0] for p in param_names], labels=[p[1] for p in param_names])
s1 = MCSamples(samples=a1l128, names=[p[0] for p in param_names], labels=[p[1] for p in param_names])
s2 = MCSamples(samples=a2l128, names=[p[0] for p in param_names], labels=[p[1] for p in param_names])
s3 = MCSamples(samples=avl128, names=[p[0] for p in param_names], labels=[p[1] for p in param_names])
```

```python
g = plots.get_subplot_plotter(subplot_size=10, width_inch = 5)

# Set line style default
g.settings.line_styles = 'tab10'
g.settings.axes_fontsize = 15
g.settings.axes_labelsize = 20
g.plot_1d([s0, s1, s2, s3], 'param1', nx=4, legend_ncol=3, lws=[3,3,3,3])
g.add_legend(legend_labels=[r'$\alpha=0$', r'$\alpha=-1$', r'$\alpha=-2$', r'$\alpha$ = Variable'])
plt.ylabel('Posterior', fontsize = 20)
plt.xlim(1.5394)
plt.axvline(x = 1.54, color = 'k', linestyle = '--', linewidth = 2.5)
# plt.title(r'Variation in $\beta_d$', fontsize = 15)
plt.savefig('plots_white/alpha_variation_betad.png', bbox_inches = 'tight')
plt.show()

g = plots.get_subplot_plotter(subplot_size=10, width_inch = 5)

# Set line style default
g.settings.line_styles = 'tab10'
g.settings.axes_fontsize = 15
g.settings.axes_labelsize = 20
g.plot_1d([s0, s1, s2, s3], 'param2', nx=4, legend_ncol=3, lws=[3,3,3,3])
plt.ylabel('Posterior', fontsize = 20)
plt.xlim(19.989,20.011)
g.add_legend(legend_labels=[r'$\alpha=0$', r'$\alpha=-1$', r'$\alpha=-2$', r'$\alpha$ = Variable'])
plt.axvline(x = 20, color = 'k', linestyle = '--', linewidth = 2.5)
# plt.title(r'Variation in $T_d$', fontsize = 15)
plt.savefig('plots_white/alpha_variation_Td.png', bbox_inches = 'tight')
plt.show()

g = plots.get_subplot_plotter(subplot_size=10, width_inch = 5)

# Set line style default
g.settings.line_styles = 'tab10'
g.settings.axes_fontsize = 15
g.settings.axes_labelsize = 20
g.plot_1d([s0, s1, s2, s3], 'param3', nx=4, legend_ncol=3, lws=[3,3,3,3])
plt.ylabel('Posterior', fontsize = 20)
g.add_legend(legend_labels=[r'$\alpha=0$', r'$\alpha=-1$', r'$\alpha=-2$', r'$\alpha$ = Variable'])
plt.axvline(x = -3, color = 'k', linestyle = '--', linewidth = 2.5)
# plt.title(r'Variation in $\beta_s$', fontsize = 15)
plt.savefig('plots_white/alpha_variation_betas.png', bbox_inches = 'tight')
plt.show()

```
